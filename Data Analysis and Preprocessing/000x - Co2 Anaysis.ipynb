```python
#************************************************************************************************************************************************** 
#In order to make a first analysis of the relationship between the power of a vehicle and its CO2 emissions I have decided to analyze the following data, which I have renamed:

#df.rename(columns={"Ewltp (g/km)": "CO2"}, inplace = True)
#df.rename(columns={"Ft": "Energy"}, inplace = True)
#df.rename(columns={"ec (cm3)": "Engine_cm3"}, inplace = True)
#df.rename(columns={"ep (KW)": "Power_KW"}, inplace = True)
#df.rename(columns={"z (Wh/km)": "Elect_Consumption"}, inplace = True)

#   the CO2 emissions are for combustion and hybrid vehicles
#   while the electric vehicles have the consumption in Watts per KM.
#   As we do not have a common information I have thought that we can equate in the following way: 
#   Apply the electricity mix (I have chosen a value of 280 grCo2/Kwh as an example) to the energy consumption.
#   to the energy consumption. before applying it I must convert the consumption that comes in watts to kilowatts, to equate all the values. 
#   (Source https://canviclimatic.gencat.cat/es/actua/factors_demissio_associats_a_lenergia/index.html)

print( '\n df_Petrol Power_KW  min-max-mean' , df_Petrol['Power_KW'].min(),' - ', df_Petrol['Power_KW'].max(),' - ', df_Petrol['Power_KW'].mean())
print( 'df_Petrol CO2 ' , df_Petrol['CO2'].mean())
    
print( '\n df_Diesel Power_KW min-max-mean' , df_Diesel['Power_KW'].min(),' - ', df_Diesel['Power_KW'].max(),' - ', df_Diesel['Power_KW'].mean())
print( 'df_Diesel CO2 ' , df_Diesel['CO2'].mean())

print( '\n df_Other Power_KW'  , df_Other['Power_KW'].min(),' - ', df_Other['Power_KW'].max(),' - ', df_Other['Power_KW'].mean())
print( 'df_Other CO2 ' , df_Other['CO2'].mean())


print( '\n df_Hibrid_Diesel Power_KW min-max-mean' , df_Petrol['Power_KW'].min(),' - ', df_Petrol['Power_KW'].max(),' - ', df_Petrol['Power_KW'].mean())

print( 'df_Hibrid_Diesel  Elect_Consumption' , df_Hibrid_Diesel['Elect_Consumption'].mean())
print( 'df_Hibrid_Diesel CO2' , df_Hibrid_Diesel['CO2'].mean() +
((df_Hibrid_Diesel['Elect_Consumption']/1000)*280).mean())

print( '\n df_Hibrid_Petrol Power_KW min-max-mean' , df_Hibrid_Petrol['Power_KW'].min(),' - ', df_Hibrid_Petrol['Power_KW'].max(),' - ', df_Hibrid_Petrol['Power_KW'].mean())
print( 'df_Hibrid_Petrol  Elect_Consumption' , df_Hibrid_Petrol['Elect_Consumption'].mean())
print( 'df_Hibrid_Petrol CO2' , df_Hibrid_Petrol['CO2'].mean() +
((df_Hibrid_Petrol['Elect_Consumption']/1000)*280).mean())

print( '\n Elect Power_KW min-max-mean' , df_Electric['Power_KW'].min(),' - ', df_Electric['Power_KW'].max(),' - ', df_Electric['Power_KW'].mean())
print( 'Elect_Consumption' , df_Electric['Elect_Consumption'].mean())
print( 'Elect_Co2' , ((df_Electric['Elect_Consumption']/1000)*280).mean())
```
