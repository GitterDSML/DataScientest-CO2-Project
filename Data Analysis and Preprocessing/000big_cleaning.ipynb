```python
#Treating Constructor NaNs Mapping from the Brand
# Count the occurrences of NaN 'Constructor' values by 'Brand'
nan_constructors = df_Ptred[df_Ptred['Constructor'].isna()]
nan_counts_by_brand = nan_constructors['Brand'].value_counts().nlargest(20)

# Get the list of brands with the top 20 highest NaN counts
top_nan_brands = nan_counts_by_brand.index.tolist()

#print(top_nan_brands)

brands_list = ['VOLVO', 'NISSAN', 'TOYOTA', 'VOLKSWAGEN', 'SKODA', 'LAND ROVER', 
               'MITSUBISHI', 'JAGUAR', 'MG', 'LEXUS', 'FORD', 'FIAT', 'MASERATI', 
               'POLESTAR', 'BENTLEY', 'FERRARI', 'SEAT', 'ASTON MARTIN', 'AIWAYS', 
               'LAMBORGHINI', 'MASERATI', 'POLESTAR', 'FERRARI', 'ASTON MARTIN', 
               'MERCEDES-BENZ', 'KIA', 'RENAULT', 'CITROEN', 'JEEP', 
               'OPEL', 'HYUNDAI', 'TESLA', 'ALFA ROMEO', 'PEUGEOT', 'MAZDA', 'PORSCHE', 'HONDA']

# Initialize an empty dictionary to store the mappings
brand_to_constructor = {}

# Loop through each brand in the list
for brand in brands_list:
    # Filter df_Ptred for the current brand where 'Constructor' is not NaN
    filtered_df = df_Ptred[(df_Ptred['Brand'] == brand) & df_Ptred['Constructor'].notna()]
    
    # Check if there are any non-NaN 'Constructor' values
    if not filtered_df.empty:
        # Find the most common 'Constructor' value for this brand
        most_common_constructor = filtered_df['Constructor'].mode()[0]
        brand_to_constructor[brand] = most_common_constructor
    else:
        # If no non-NaN 'Constructor' values are found, map the brand to None or some default
        brand_to_constructor[brand] = None

# Output the resulting dictionary
#print(brand_to_constructor)

# First, create a mapping series from the brand to constructor
brand_constructor_mapping = pd.Series(brand_to_constructor)

# Use 'map' to create a Series of constructors mapped from the 'Brand' column
mapped_constructors = df_Ptred['Brand'].map(brand_constructor_mapping)

# Now fill NaN values in the 'Constructor' column using this mapped Series
df_Ptred['Constructor'] = df_Ptred['Constructor'].fillna(mapped_constructors)

# After attempting to fill with mapped constructors, fill remaining NaNs with the 'Brand' name
df_Ptred['Constructor'] = df_Ptred['Constructor'].fillna(df_Ptred['Brand'])

# If you want to double-check or see the changes, you can display the DataFrame

df_Ptred.info()
```

```python
# Function to apply to each row for filling 'Version'
def fill_version(row):
    if pd.isna(row['Version']):
        # Try to find a matching 'Version' for the same 'Brand', 'Constructor', 'Year', and 'Energy'
        match = df_Ptred[(df_Ptred['Brand'] == row['Brand']) &
                         (df_Ptred['Constructor'] == row['Constructor']) &
                         (df_Ptred['year'] == row['year']) &
                         (df_Ptred['Energy'] == row['Energy']) &
                         ~pd.isna(df_Ptred['Version'])]
        
        if not match.empty:
            return match['Version'].iloc[0]  # Return the first match found
        else:
            # Fetch the most common 'Version' for the specific 'Brand', 'Energy', and 'Year'
            common_version = df_Ptred[(df_Ptred['Brand'] == row['Brand']) &
                                      (df_Ptred['Energy'] == row['Energy']) &
                                      (df_Ptred['year'] == row['year'])]['Version'].mode()
            if not common_version.empty:
                return common_version.iloc[0]  # Return the most common 'Version'
            else:
                return pd.NA  # Preserve NaN if no version is found
    else:
        return row['Version']

# Apply the function to fill in 'Version'   +++++++++++++++++++++++++++++++++++++  Marta: In my opinion and knowledge this is not useful. and overload the process

# df_Ptred['Version'] = df_Ptred.apply(fill_version, axis=1)
```

```python
# List of columns to fill NaN values with mean
columns_to_fill = ['Test_mass', 'Wheelbase_mm', 'Axle_width_steer_mm', 'Axle_width_other_mm']

# Grouping criteria
grouping_columns = ['Veh_Model']

# Applying the transformation
for column in columns_to_fill:
    # Compute the mean for each group and fill NaN values in the column
    df_Ptred[column] = df_Ptred.groupby(grouping_columns)[column].transform(lambda x: x.fillna(x.mean()))

# Check the results to ensure NaN values are filled
print(df_Ptred[columns_to_fill].isna().sum())

# Remove rows that still contain NaN values in any of the four specified columns
df_Ptred.dropna(subset=columns_to_fill, inplace=True)

# Verify the operation by checking for NaN values again
print("After cleaning:")
print(df_Ptred[columns_to_fill].isna().sum())
```

```python
#NaN treatment 'Power_KW', 'El_Consumpt_whkm', 'Fuel consumption', 'Electric range (km)', 'Elect_Consumption', 'Erwltp (g/km) 
## averaging based on model

# Define grouping criteria
grouping_columns = ['Veh_Model', 'Version']
grouping_columns2 = ['Brand', 'Veh_type','Energy']
grouping_columns3 = ['Energy']
grouping_columns4 = ['Veh_type','Energy']

# Treat NaNs for 'Power_KW'
df_Ptred.loc[df_Ptred['Energy'] == 'electric', 'Power_KW'] = df_Ptred[df_Ptred['Energy'] == 'electric'].groupby(grouping_columns4)['Power_KW'].transform(lambda x: x.fillna(x.mean()))

# Treat NaNs for 'El_Consumpt_whkm'
df_Ptred['El_Consumpt_whkm'] = np.where(df_Ptred['Energy'].isin(['diesel', 'petrol', 'lpg']),
                                        df_Ptred['El_Consumpt_whkm'].fillna(0),
                                        df_Ptred.groupby('Veh_Model')['El_Consumpt_whkm'].transform(lambda x: x.fillna(x.mean())))

# Treat NaNs for 'Fuel consumption'
df_Ptred['Fuel consumption'] = np.where(df_Ptred['Energy'] == 'electric',
                                        df_Ptred['Fuel consumption'].fillna(0),
                                        df_Ptred.groupby(grouping_columns4)['Fuel consumption'].transform(lambda x: x.fillna(x.mean())))

# Treat NaNs for 'Electric range (km)'
df_Ptred['Electric range (km)'] = np.where(df_Ptred['Energy'].isin(['diesel', 'petrol', 'lpg']),
                                          df_Ptred['Electric range (km)'].fillna(0),
                                          df_Ptred.groupby(grouping_columns2)['Electric range (km)'].transform(lambda x: x.fillna(x.mean())))

# Treat NaNs for 'Elect_Consumption'
df_Ptred['Elect_Consumption'] = np.where(df_Ptred['Energy'].isin(['diesel', 'petrol', 'lpg']),
                                         df_Ptred['Elect_Consumption'].fillna(0),
                                         df_Ptred.groupby(grouping_columns)['Elect_Consumption'].transform(lambda x: x.fillna(x.mean())))

# Treat NaNs for 'Erwltp (g/km)'
df_Ptred['Erwltp (g/km)'] = df_Ptred['Erwltp (g/km)'].fillna(0)

# Treat NaNs for 'Engine_cm3'
df_Ptred['Engine_cm3'] = df_Ptred['Engine_cm3'].fillna(0)

# Check if any NaNs remain and remove rows where any of these columns still have NaN
df_Ptred.dropna(subset=['Power_KW', 'El_Consumpt_whkm', 'Fuel consumption', 'Electric range (km)', 'Elect_Consumption', 'Erwltp (g/km)', 'Engine_cm3', 'Elect_Consumption'], inplace=True)

# Verify that no NaN values are left in the specified columns

# Check the results
print(df_Ptred[['Power_KW', 'El_Consumpt_whkm', 'Fuel consumption', 'Electric range (km)', 'Elect_Consumption', 'Erwltp (g/km)' , 'Engine_cm3']].isna().sum())
```

```python
### 
def assign_labels(bin_range):
    """ Helper function to format bin range labels. """
    if pd.isnull(bin_range.right):
        return f"{round(bin_range.left)}+ kw"
    return f"{round(bin_range.left)} - {round(bin_range.right)} kw"

def create_pivot_table(df, param_x, param_y, value_column, num_ranges):
    """
    Creates a pivot table with specified parameters.
    
    Args:
    df (DataFrame): The DataFrame to operate on.
    param_x (list): Column names to use as columns in the pivot table.
    param_y (str): Column name to use for creating range bins and rows in the pivot table.
    value_column (str): Column name whose data is to be summarized.
    num_ranges (int): Number of quantile ranges to divide the data into.

    Returns:
    DataFrame: A pivot table based on the inputs.
    """
    # Create quantile bins
    df['Range'] = pd.qcut(df[param_y], num_ranges, duplicates='drop')
    df['Label'] = df['Range'].apply(assign_labels)
    
    # Create pivot table
    pivot_table = pd.pivot_table(df, values=value_column, index='Label', columns=param_x, aggfunc='sum', fill_value=0)
    
    return pivot_table

#  usage
param_x = ['Energy']  # Specify based on case
param_y = 'Power_KW'
value_column = 'Engine_cm3'
num_ranges = 4
pivot_name = 'Pivot_' + '_'.join(param_x) + '_' + param_y

print("Your new table name is:", pivot_name)

# Call the function
pivot_table_result = create_pivot_table(df_Ptred, param_x, param_y, value_column, num_ranges)

# Create a DataFrame from the pivot table data for display
df_table = pd.DataFrame(pivot_table_result.to_records())
df_table = df_table.reset_index(drop=True)
display(df_table.head())
```

```python
# Review the distribution of  WLTP and NDEC CO2 indicators
# Prepare the data for the violin plot
# Melt the DataFrame to long-format for seaborn compatibility
melted_df = df_Ptclean.melt(value_vars=['CO2_wltp', 'Enedc (g/km)'], var_name='Metric', value_name='Value')

# Create the violin plot
plt.figure(figsize=(6, 15))
violin = sns.violinplot(x='Metric', y='Value', data=melted_df, split=True, inner=None, palette='muted')

# Calculate and annotate metrics for each distribution
for i, metric in enumerate(['CO2_wltp', 'Enedc (g/km)']):
    # Extract the data for the current metric
    data = df_Ptred[metric].dropna()
    
    # Calculate metrics
    mean = data.mean()
    median = data.median()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    
    # Annotate the violin plot with these metrics
    plt.text(i, data.min(), f'Min: {data.min():.2f}', ha='center')
    plt.text(i, q1, f'Q1: {q1:.2f}', ha='center')
    plt.text(i, median, f'Median: {median:.2f}', ha='center')
    plt.text(i, q3, f'Q3: {q3:.2f}', ha='center')
    plt.text(i, data.max(), f'Max: {data.max():.2f}', ha='center')
    plt.text(i, mean, f'Mean: {mean:.2f}', va='center', ha='center', color='white')

# Finalize the plot
plt.title('Distribution Comparison of CO2_wltp and Enedc (g/km)')
plt.ylabel('Values')
plt.xticks(np.arange(2), ['CO2_wltp', 'Enedc (g/km)'])

plt.show()
```
